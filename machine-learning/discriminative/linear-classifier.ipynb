```python
# dosya okuma
file_path = "red-black.data"
f = open(file_path, "r")
lines = f.read().splitlines()
num_row = len(lines) - 1
for line in lines:
    if line.startswith('#'):
        num_col = len(line[1:].split('\t'))
        break
num_feat = num_col - 1
```

```python
# data tablosunun oluşturulması
data = []
set_label = set()
for line in lines:
    if line.startswith('#'):
        title = line[1:].split('\t')
        continue
    values = line.split('\t')
    set_label.add(values[-1])
    for i in range(len(values) - 1):
        values[i] = int(values[i]) # son değer hariç tümü int tipine dönüştürülür
    data.append(values)
list_label = sorted(list(set_label), reverse = True)
print("data >> ", data)
print("\nlist_label >> ", list_label)
```

```python
def predict(input, weights):
    sum = 0
    for i in range(len(weights) - 1):
        sum += input[i] * weights[i]
    sum += weights[-1]
    if sum < 0:
        return list_label[0]
    else:
        return list_label[1]
```

```python
# tekil test
print(predict([4, 4],[2,-1,-5]))
```

```python
# eğitim verisi üzerinde kaba kuvvet yöntemiyle olası bütün doğru denklemlerinin taranması
# ve veri tablosunu hatasız şekilde sınıflandıran doğruların listelenmesi
for w1 in range(-10, 10):
    for w2 in range(-10, 10):
        for w0 in range(-10, 10):
            c = 0
            for row in data:
                if predict(row[:-1], [w1, w2, w0]) == row[-1]:
                    c += 1
            accuracy = c / len(data)
            if accuracy == 1.0:
                print('[{0}, {1}, {2}]'.format(w1, w2, w0))
```
