```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
# Load the data table into a pandas DataFrame
data = pd.DataFrame({
    'salary': ['low', 'low', 'low', 'low', 'low', 'average', 'average', 'average', 'average', 'average', 'high', 'high', 'high', 'high', 'high'],
    'age': ['young', 'middle', 'middle', 'old', 'old', 'young', 'young', 'middle', 'old', 'old', 'young', 'young', 'middle', 'middle', 'old'],
    'debt': ['indebted', 'debtless', 'indebted', 'debtless', 'indebted', 'debtless', 'indebted', 'debtless', 'debtless', 'indebted', 'debtless', 'indebted', 'debtless', 'indebted', 'debtless'],
    'credit': ['no', 'yes', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes']
})
```

```python
# Map categorical features to numerical values
data['salary'] = data['salary'].map({'low': 0, 'average': 1, 'high': 2})
data['age'] = data['age'].map({'young': 0, 'middle': 1, 'old': 2})
data['debt'] = data['debt'].map({'debtless': 0, 'indebted': 1})
```

```python
# Split the data into input features and output variable
X = data[['salary', 'age', 'debt']]
y = data['credit']
```

```python
# Train a decision tree classifier with the entropy criterion
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)
```

```python
# Visualize the decision tree
tree.plot_tree(clf)
```
